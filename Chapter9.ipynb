### 9.1 - Dictionaries

```python
#Diferencia entre colecciones

#Listas
#Colleción lineal de elementos que deben estar en orden

#Diccionarios
#Una "bolsa" de valores, cada uno con su propia etiqueta
#Las listas indexan los elementos basados en su posición
#Los diccionarios indexan los elementos basados en sus etiquetas

purse = dict()
#purse['key'] = value
purse['money'] = 12
purse['candy'] = 3
purse['tissues'] = 75
print(purse)

#Imprimir el valor que almacena una llave
print(purse['candy'])

#Modificar el valor de una llave
purse['candy'] = purse['candy'] + 2 
print(purse)

#Podemos crar un diccionario vacío usando llaves
ooo = {}
print(ooo)
```

### 9.2 - Counting with Dictionaries

```python
#Most common Name?

#Un uso común de los diccionarios es 
#contar cuántas veces vemos "algo"

names = dict()
names['Mark'] = 1
names['Matt'] = 1
print(names)

#Encontramos otro Mark
names['Mark'] = names['Mark'] + 1
print(names)
```

```python
#Dictionary Tracebacks

#Si intentamos buscar una llave que no existe
#Python nos arrojará un error
#Podemos usar el operador "in" para ver si una llave existe en un diccionario

forTracebacks = dict()
print(forTracebacks['zzz'])
```

```python
#Manera correcta
'zzz' in forTracebacks
```

```python
#When we see a new name

#Cuando vemos un nombre nuevo
#Tenemos que agregarlo al diccionario
#Si el nombre ya existe, tenemos que incrementar el contador

counts = dict()
nombres = ['csev','cwen','csev','zqian','cwen']
for nombre in nombres:
    if nombre not in counts:
        counts[nombre] = 1
    else:
        counts[nombre] = counts[nombre] + 1
print(counts)
```

```python
#The get method for dictionaries

#El patrón de buscar una llave y agregarla si no existe
#es tan común que hay un método llamado get() que hace esto por nosotros

if nombre in counts:
    x = counts[nombre]
else:
    x = 0

#Podemos usar get() y proveer un valor por defecto de 0 cuando la llave no existe
x = counts.get(nombre,0)

print(x)
```

```python
#Simplified counting with get()

counts2 = dict()
names2 = ['csev','cwen','csev','zqian','cwen']

for name in names2:
    counts2[name] = counts2.get(name,0) + 1
print(counts2)
```

### 9.3 - Dictionaries and Files

```python
#Counting pattern

countlines = dict()
print('Enter a line of text:')
linetocount = input('')

palabras = linetocount.split()

print('Words:', palabras)

print('Counting...')
for palabra in palabras:
    countlines[palabra] = countlines.get(palabra,0) + 1
print('Counts', countlines)
```

```python
#Definite loops and dictionaries

#Podemos escribir un bucle definido para iterar a través de las llaves de un diccionario

counts3 = {'chuck':1, 'fred':42, 'jan':100}
for key in counts3:
    print(key, counts3[key])

#Retrieving lists of Keys and Values
print('')
#Podemos obtener una lista de las llaves, los valores o ambos

jjj = {'chuck':1, 'fred':42, 'jan':100}
print("jjj:",list(jjj))
#Esto imprime una lista de las llaves

print(jjj.keys())

print(jjj.values())

print(jjj.items())

#Two iteration variables

#Podemos usar dos iteradores para iterar a través de las llaves 
# y los valores de un diccionario
print('')
for aaa,bbb in jjj.items():
    print(aaa,bbb)
```

```python
#Leer archivo

#namefile = input('Enter file: ')
#handle = open(namefile)
handle = open('docs/chapter9file.txt')

wordcount = dict()
for line in handle:
    words = line.split()
    for word in words:
        wordcount[word] = wordcount.get(word,0) + 1

bigcount = None
bigword = None

for word,count in wordcount.items():
    if bigcount is None or count > bigcount:
        bigword = word
        bigcount = count

print("The most common word is:", bigword, "with", bigcount, "times")
```
